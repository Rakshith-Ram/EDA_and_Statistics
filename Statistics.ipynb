# **Q-Q plots**

To find if the random variable is of a Gaussian/Normal distribution or not

```python
import numpy as np
import scipy.stats as stats
import pylab
```

```python
# Generating 1000 samples from N(0,1)

std_norm = np.random.normal(loc=0, scale=1, size=1000)
```

```python
# Percentiles of std_norm at intervals of 5 percentiles, from 0th to 100th percentile

for i in range(0,101,5):
  print(i, np.percentile(std_norm,i))
```

```python
# Generating 1000 samples of normal/gaussian distribution

sample = np.random.normal(loc=20, scale=5, size=1000)

# try with  100, 10000 etc.
```

```python
# To generate a QQ plot :-

stats.probplot(sample, dist="norm", plot=pylab)
pylab.show()
```

```python
# Generating 1000 samples of uniform distribution

sample = np.random.uniform(low=-5, high=5, size=1000)
```

```python
# To generate a QQ plot :-

stats.probplot(sample, dist="norm", plot=pylab)
pylab.show()
```

# **Box cox transform**

Convert any distribution to Normal/Gaussian distribution

```python
import numpy as np
import pylab
import random
import scipy.stats as stats
```

```python
sample = np.random.uniform(low=0, high=1, size=1000)

stats.probplot(sample, dist="norm", plot=pylab)
pylab.show()
```

```python
sample_transformed, _ = stats.boxcox(sample)

stats.probplot(sample_transformed, dist="norm", plot=pylab)
pylab.show()
```

```python
sample = np.random.lognormal(size=1000)

stats.probplot(sample, dist="norm", plot=pylab)
pylab.show()
```

```python
sample_transformed, _ = stats.boxcox(sample)

stats.probplot(sample_transformed, dist="norm", plot=pylab)
pylab.show()
```
