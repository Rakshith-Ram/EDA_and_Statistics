```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Iris dataset

```python
iris_data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Statistics+EDA/Iris.csv")
```

## 2D scatter plots

```python
sns.scatterplot(x=iris_data["PetalLengthCm"], y=iris_data["PetalWidthCm"], hue=iris_data["Species"])
```

```python
sns.scatterplot(iris_data, x="SepalLengthCm", y="SepalWidthCm", hue="Species")
```

```python
sns.histplot(iris_data, x="PetalLengthCm", hue="Species", bins=50, kde=True)
```


## Pair plots

```python
sns.pairplot(iris_data, hue="Species")
```

```python
iris_setosa = iris_data[iris_data["Species"] == "Iris-setosa"]
iris_versicolor = iris_data[iris_data["Species"] == "Iris-versicolor"]
iris_virginica = iris_data[iris_data["Species"] == "Iris-virginica"]
```

```python
iris_setosa.head()
```

## Mean, Varience and Standard deviation

```python
print(np.mean(iris_setosa["PetalLengthCm"]))
print(np.mean(iris_versicolor["PetalLengthCm"]))
print(np.mean(iris_virginica["PetalLengthCm"]))
```

```python
print(np.var(iris_setosa["PetalLengthCm"]))
print(np.mean(iris_versicolor["PetalLengthCm"]))
print(np.mean(iris_virginica["PetalLengthCm"]))
```

## Box plots

```python
boxplot = sns.boxplot(data=iris_data, x="PetalLengthCm", y="Species", hue="Species")
boxplot.set_title("Boxplot of Petal Length by Species")
plt.show()
```

```python
violinplot = sns.violinplot(data=iris_data, x="PetalLengthCm", y="Species", hue="Species")
violinplot.set_title("Violinplot of Petal Length by Species")
plt.show()
```
